```python
#reading csv file
import pandas as pd
headers_name = ['First_Term_Gpa','Second_Term_Gpa', 'First_Language', 'Funding', 'School', 'FastTrack', 'Coop','Residency','Gender','Previous_Education','Age_Group','High_School_Average_Mark','Math_Score','English_Grade','First_Year_Persistence']
dataset = pd.read_csv('Student data.csv', names=headers_name) 
dataset.head(50)

```

```python
#dropping the first 23 rows
dataset.drop(dataset.index[:24], inplace=True)
dataset.head(50)
```

```python
dataset.info()
```

```python
#checking for missing values
print("# of values of '?' in First_Term_Gpa:",dataset['First_Term_Gpa'].where(dataset['First_Term_Gpa'] == '?').count())
print("# of values of '?' in Second_Term_Gpa:",dataset['Second_Term_Gpa'].where(dataset['Second_Term_Gpa'] == '?').count())
for column in dataset.columns:
    print(column, dataset[column].unique())

```

```python
import numpy as np
#replacing the '?' with NaN
dataset_cleaned = dataset.replace('?', np.nan)
print("# of values of '?' in First_Term_Gpa",dataset_cleaned['First_Term_Gpa'].where(dataset_cleaned['First_Term_Gpa'] == '?').count())
print("# of values of '?' in Second_Term_Gpa",dataset_cleaned['Second_Term_Gpa'].where(dataset_cleaned['Second_Term_Gpa'] == '?').count())

for column in dataset_cleaned.columns:
    print(column, dataset_cleaned[column].unique())
```

```python
dataset_cleaned.info()
```

```python
#Because all columns has numerical values, they will be converted to float32
dataset_cleaned = dataset_cleaned.astype('float32')
print(dataset_cleaned.info())
dataset_cleaned.head(3)
```

```python
#dropping the null values from First_Term_Gpa
dataset_cleaned.dropna(subset=['First_Term_Gpa'], inplace=True)
dataset_cleaned.isnull().sum()
```

```python
dataset_cleaned.drop('High_School_Average_Mark', axis=1, inplace=True)
dataset_cleaned.isnull().sum()
```

```python
dataset_cleaned.describe()
```

```python
#completing the missing values of English_Grade with the median
columns_4median = ['English_Grade']
from sklearn.impute import SimpleImputer
imputer_SecGPA_Math_Eng = SimpleImputer(strategy='median')
imputer_SecGPA_Math_Eng.fit(dataset_cleaned[columns_4median])
dataset_cleaned[columns_4median] = imputer_SecGPA_Math_Eng.transform(dataset_cleaned[columns_4median])
dataset_cleaned.isnull().sum()
```

```python
imputer_FirstLanguage = SimpleImputer(strategy='most_frequent')
imputer_FirstLanguage.fit(dataset_cleaned[['First_Language']])
dataset_cleaned[['First_Language']] = imputer_FirstLanguage.transform(dataset_cleaned[['First_Language']])
dataset_cleaned.isnull().sum()
```

```python
from sklearn.impute import KNNImputer

knn_imputer = KNNImputer(n_neighbors=4)
dataset_cleaned = pd.DataFrame(knn_imputer.fit_transform(dataset_cleaned),columns = dataset_cleaned.columns)
dataset_cleaned.isnull().sum()
```

```python
#plotting the histogram of the dataset
import matplotlib.pyplot as plt
dataset_cleaned.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
#plotting the correlation between the columns
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,10))
sns.heatmap(dataset_cleaned.corr(), annot=True, cmap='coolwarm')
```

```python
def preprocessingdata(data):
    categorical_columns = ['First_Language','Funding','Gender']
    #applying one hot encoding to the categorical columns
    data = pd.get_dummies(data, columns=categorical_columns, dtype='float32') 
    data['FastTrack'] = data['FastTrack'].replace({1: 0, 2: 1})
    data['Coop'] = data['Coop'].replace({1: 0, 2: 1})
    data['Residency'] = data['Residency'].replace({1: 0, 2: 1})
    return data

dataset_cleaned = preprocessingdata(dataset_cleaned)
dataset_cleaned.head(3)

```

```python
for column in dataset_cleaned.columns:
    print(column, dataset_cleaned[column].unique())
```

```python
#reviewing and treating the missing values
print(dataset_cleaned.isnull().sum())
dataset_cleaned.describe()

```

```python
X = dataset_cleaned.drop('First_Year_Persistence', axis=1)
y = dataset_cleaned['First_Year_Persistence']

#splitting the dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42, test_size=0.2, stratify=y)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#scaling the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train_cleaned = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_cleaned = pd.DataFrame(X_test_scaled, columns=X_test.columns)
X_train_cleaned.info()
```
