```python
#reading csv file
import pandas as pd
headers_name = ['First_Term_Gpa','Second_Term_Gpa', 'First_Language', 'Funding', 'School', 'FastTrack', 'Coop','Residency','Gender','Previous_Education','Age_Group','High_School_Average_Mark','Math_Score','English_Grade','First_Year_Persistence']
dataset = pd.read_csv('Student data.csv', names=headers_name) 
dataset.head(50)

```

```python
#dropping the first 23 rows
dataset.drop(dataset.index[:24], inplace=True)
dataset.head(50)
```

```python
dataset.info()
```

```python
#checking for missing values
print("# of values of '?' in First_Term_Gpa:",dataset['First_Term_Gpa'].where(dataset['First_Term_Gpa'] == '?').count())
print("# of values of '?' in Second_Term_Gpa:",dataset['Second_Term_Gpa'].where(dataset['Second_Term_Gpa'] == '?').count())
for column in dataset.columns:
    print(column, dataset[column].unique())

```

```python
import numpy as np
#replacing the '?' with NaN
dataset_cleaned = dataset.replace('?', np.nan)
print("# of values of '?' in First_Term_Gpa",dataset_cleaned['First_Term_Gpa'].where(dataset_cleaned['First_Term_Gpa'] == '?').count())
print("# of values of '?' in Second_Term_Gpa",dataset_cleaned['Second_Term_Gpa'].where(dataset_cleaned['Second_Term_Gpa'] == '?').count())

for column in dataset_cleaned.columns:
    print(column, dataset_cleaned[column].unique())
```

```python
dataset_cleaned.info()
```

```python
#Because all columns has numerical values, they will be converted to float
dataset_cleaned = dataset_cleaned.astype(float)
print(dataset_cleaned.info())
dataset_cleaned.head(3)
```

```python
#completing the missing values of English_Grade with the median
from sklearn.impute import SimpleImputer
imputer_SecGPA_Math_Eng = SimpleImputer(strategy='median')
imputer_SecGPA_Math_Eng.fit(dataset_cleaned[['Second_Term_Gpa','Math_Score','English_Grade']])
dataset_cleaned[['Second_Term_Gpa','Math_Score','English_Grade']] = imputer_SecGPA_Math_Eng.transform(dataset_cleaned[['Second_Term_Gpa','Math_Score','English_Grade']])
dataset_cleaned.isnull().sum()
```

```python
imputer_FirstLanguage = SimpleImputer(strategy='most_frequent')
imputer_FirstLanguage.fit(dataset_cleaned[['First_Language']])
dataset_cleaned[['First_Language']] = imputer_FirstLanguage.transform(dataset_cleaned[['First_Language']])
dataset_cleaned.isnull().sum()
```
