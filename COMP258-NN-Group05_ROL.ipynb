```python
#reading csv file
import pandas as pd
headers_name = ['First_Term_Gpa','Second_Term_Gpa', 'First_Language', 'Funding', 'School', 'FastTrack', 'Coop','Residency','Gender','Previous_Education','Age_Group','High_School_Average_Mark','Math_Score','English_Grade','First_Year_Persistence']
df = pd.read_csv('Student data.csv', names=headers_name) 
df.head(50)

```

```python
#dropping the first 23 rows
df.drop(df.index[:24], inplace=True)
df.head(50)
```

```python
df.info()
```

```python
#splitting the data into training and testing
from sklearn.model_selection import train_test_split
X = df.drop('First_Year_Persistence',axis=1)
y = df['First_Year_Persistence']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify = y)
```

```python
#printing unique values of each column
#print("Unique values of First_Term_Gpa:",df['First_Term_Gpa'].unique())
#print("Unique values of Second_Term_Gpa:",df['Second_Term_Gpa'].unique())
print("# of values of '?' in First_Term_Gpa:",X_train['First_Term_Gpa'].where(X_train['First_Term_Gpa'] == '?').count())
print("# of values of '?' in Second_Term_Gpa:",X_train['Second_Term_Gpa'].where(X_train['Second_Term_Gpa'] == '?').count())
print("Unique values of First_Language:",X_train['First_Language'].unique())
print("Unique values of Funding:",X_train['Funding'].unique())
print("Unique values of School:",X_train['School'].unique())
print("Unique values of FastTrack:",X_train['FastTrack'].unique())
print("Unique values of Coop:",X_train['Coop'].unique())
print("Unique values of Residency:",X_train['Residency'].unique())
print("Unique values of Gender:",X_train['Gender'].unique())
print("Unique values of Previous_Education:",X_train['Previous_Education'].unique())
print("Unique values of Age_Group:",X_train['Age_Group'].unique())
print("Unique values of High_School_Average_Mark:",X_train['High_School_Average_Mark'].unique())
print("Unique values of Math_Score:",X_train['Math_Score'].unique())
print("Unique values of English_Grade:",X_train['English_Grade'].unique())
print("Unique values of First_Year_Persistence:",y_train.unique())

```

```python
import numpy as np
#replacing the '?' with NaN
X_train_cleaned = X_train.replace('?', np.nan)
print("# of values of '?' in First_Term_Gpa",X_train_cleaned['First_Term_Gpa'].where(X_train_cleaned['First_Term_Gpa'] == '?').count())
print("# of values of '?' in Second_Term_Gpa",X_train_cleaned['Second_Term_Gpa'].where(X_train_cleaned['Second_Term_Gpa'] == '?').count())
print("Unique values of First_Language:",X_train_cleaned['First_Language'].unique())
print("Unique values of Funding:",X_train_cleaned['Funding'].unique())
print("Unique values of School:",X_train_cleaned['School'].unique())
print("Unique values of FastTrack:",X_train_cleaned['FastTrack'].unique())
print("Unique values of Coop:",X_train_cleaned['Coop'].unique())
print("Unique values of Residency:",X_train_cleaned['Residency'].unique())
print("Unique values of Gender:",X_train_cleaned['Gender'].unique())
print("Unique values of Previous_Education:",X_train_cleaned['Previous_Education'].unique())
print("Unique values of Age_Group:",X_train_cleaned['Age_Group'].unique())
print("Unique values of High_School_Average_Mark:",X_train_cleaned['High_School_Average_Mark'].unique())
print("Unique values of Math_Score:",X_train_cleaned['Math_Score'].unique())
print("Unique values of English_Grade:",X_train_cleaned['English_Grade'].unique())
```

```python
X_train_cleaned.info()
```

```python
#replacing the values of FastTrack column 2 with 1 and 1 with 0
X_train_cleaned['FastTrack'].replace(1, 0, inplace=True)
X_train_cleaned['FastTrack'].replace(2, 1, inplace=True)
X_train_cleaned['Coop'].replace(1, 0, inplace=True)
X_train_cleaned['Coop'].replace(2, 1, inplace=True)
X_train_cleaned['Residency'].replace(1, 0, inplace=True)
X_train_cleaned['Residency'].replace(2, 1, inplace=True)
print("Unique values of FastTrack:",X_train_cleaned['FastTrack'].unique())
print("Unique values of Coop:",X_train_cleaned['Coop'].unique())
print("Unique values of Residency:",X_train_cleaned['Residency'].unique())
```

```python
#defining numerical and categorical columns
numerical_columns = ['First_Term_Gpa','Second_Term_Gpa','High_School_Average_Mark','Math_Score']
categorical_columns = ['First Language', 'Funding', 'School', 'FastTrack','Coop', 'Residensy', 'Gender', 'Previous_Education', 'Age_Group', 'English_Grade', 'First_Year_Persistence']

#converting the numerical columns to double
X_train_cleaned[numerical_columns] = X_train_cleaned[numerical_columns].astype(float)
print(X_train_cleaned.info())
X_train_cleaned.head(3)
```

```python
#reviewing and treating the missing values
print(X_train_cleaned.isnull().sum())
X_train_cleaned.describe()

```

```python
X_train_cleaned[pd.isnull(X_train_cleaned['First_Term_Gpa'])]
```

```python
X_train_cleaned = X_train_cleaned.dropna(subset=['First_Term_Gpa'])
X_train_cleaned.info()
```

```python
print(X_train_cleaned.shape)
X_train_cleaned.describe()
```

```python

X_train_cleaned['Second_Term_Gpa'] = X_train_cleaned['Second_Term_Gpa'].fillna(X_train_cleaned['Second_Term_Gpa'].mean())
X_train_cleaned.isnull().sum()
```

```python
X_train_cleaned['First_Language'] = X_train_cleaned['First_Language'].fillna('4')
X_train_cleaned.isnull().sum()
```

```python
#counting the unique values of column Age_Group
print(X_train_cleaned['Age_Group'].value_counts())
print(X_train_cleaned['Previous_Education'].value_counts())
#adding the mode value of Age_Group to the missing values
X_train_cleaned['Age_Group'] = X_train_cleaned['Age_Group'].fillna('3')

#adding the mode value of Previous_Education to the missing values
X_train_cleaned['Previous_Education'] = X_train_cleaned['Previous_Education'].fillna('1')

X_train_cleaned.isnull().sum()
```

```python
#dropping the column 'High_School_Average_Mark' as it has too many missing values
X_train_cleaned.drop('High_School_Average_Mark', axis=1, inplace=True)

#dropping the column 'School' as it only has one value
X_train_cleaned.drop('School', axis=1, inplace=True)

X_train_cleaned.isnull().sum()
```

```python
#completing the missing values of column 'Math_Score' with the mean value
X_train_cleaned['Math_Score'] = X_train_cleaned['Math_Score'].fillna(X_train_cleaned['Math_Score'].mean())
X_train_cleaned.isnull().sum()
```

```python
#counting the unique values of column 'English_Grade'
print(X_train_cleaned['English_Grade'].value_counts())

#adding the 5th value to the missing values of column 'English_Grade'. Creating a new category '5'
X_train_cleaned['English_Grade'] = X_train_cleaned['English_Grade'].fillna('5')
print(X_train_cleaned['English_Grade'].value_counts())
```

```python
#converting the columns 'Previos_Education', 'Age_Group' and 'English_Grade' to float because they can influence the model by the weight of the numbers
X_train_cleaned['Previous_Education'] = X_train_cleaned['Previous_Education'].astype(float)
X_train_cleaned['Age_Group'] = X_train_cleaned['Age_Group'].astype(float)
X_train_cleaned['English_Grade'] = X_train_cleaned['English_Grade'].astype(float)
X_train_cleaned.info()
```
